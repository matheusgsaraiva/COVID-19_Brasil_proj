# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Análise dos Dados

**Dicionário das variáveis**

Veja abaixo o nome e o significado das colunas do conjunto de dados utilizados:

* **iso_code** - ISO 3166-1 alfa-3 - códigos de países de três letras (Ex: Brasil fica BRA);
* **continent** - Continente da localização geográfica;
* **location** - Localização geográfica;
* **date** - Data de observação;
* **total_cases** - Total de casos confirmados de COVID-19. As contagens podem incluir casos prováveis, quando relatados;
* **new_cases** - Novos casos confirmados de COVID-19. As contagens podem incluir casos prováveis, onde foram relatados. Em casos raros em que a fonte relata uma mudança diária negativa devido a uma correção de dados, define-se esta métrica para NA;
* **new_cases_smoothed** - Novos casos confirmados de COVID-19 (7 dias suavizados). As contagens podem incluir casos prováveis, onde são relatados;
* **total_deaths** - Total de mortes atribuídas à COVID-19. As contagens podem incluir prováveis mortes, quando relatadas;
* **new_deaths** - Novos óbitos atribuídos à COVID-19. As contagens podem incluir mortes prováveis, quando relatadas. Em casos raros onde a fonte relata uma mudança diária negativa devido a uma correção de dados, define-se esta métrica para NA;
* **new_deaths_smoothed** - Novos óbitos atribuídos à COVID-19 (7 dias suavizados). As contagens podem incluir prováveis mortes, quando relatadas;
* total_cases_per_million - Total de casos confirmados de COVID-19 por 1 milhão de pessoas. As contagens podem incluir casos prováveis, onde foram relatados;
* **new_cases_per_million** - Novos casos confirmados de COVID-19 por 1 milhão de pessoas. As contagens podem incluir casos prováveis, quando relatados;
* **new_cases_smoothed_per_million** - Novos casos confirmados de COVID-19 (7 dias suavizados) por 1 milhão de pessoas. As contagens podem incluir casos prováveis, quando relatados;
* **total_deaths_per_million** - Total de mortes atribuídas à COVID-19 por 1 milhão de pessoas. As contagens podem incluir mortes prováveis, quando relatadas;
* **new_deaths_per_million** - Novas mortes atribuídas à COVID-19 por 1 milhão de pessoas. As contagens podem incluir prováveis mortes, quando relatadas;
* **new_deaths_smoothed_per_million** - Novas mortes atribuídas à COVID-19 (7 dias suavizados) por 1 milhão de pessoas. As contagens podem incluir prováveis mortes, quando relatadas;
* **reproduction_rate** - Estimativa em tempo real da taxa efetiva de reprodução (R) da COVID-19. Ver https://github.com/crondonm/TrackingR/tree/main/Estimates-Database;
* **icu_patients** - Número de pacientes COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia;
* **icu_patients_per_million** - Número de pacientes COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia por 1 milhão de pessoas;
* **hosp_patients** - Número de pacientes da COVID-19 no hospital em um determinado dia;
* **hosp_patients_per_million** - Número de pacientes da COVID-19 no hospital em um determinado dia por 1 milhão de pessoas;
* **weekly_icu_admissions** - Número de pacientes COVID-19 recém admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana (data do relatório e os 6 dias anteriores);
* **weekly_icu_admissions_per_million** - Número de pacientes COVID-19 recém admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana por 1 milhão de pessoas (data do relatório e os 6 dias anteriores);
* **weekly_hosp_admissions** - Número de pacientes COVID-19 recém admitidos em hospitais em uma determinada semana (data do relatório e os 6 dias anteriores);
* **weekly_hosp_admissions_per_million** - Número de pacientes COVID-19 recém admitidos em hospitais em uma determinada semana por 1 milhão de pessoas (data do relatório e os 6 dias anteriores);
* **total_tests** - Testes totais para a COVID-19;
* **new_tests** - Novos testes para a COVID-19 (apenas calculados para dias consecutivos);
* **total_tests_per_thousand** - Testes totais para COVID-19 por 1.000 pessoas;
* **new_tests_per_thousand** - Novos testes para a COVID-19 por 1.000 pessoas;
* **new_tests_smoothed** - Novos testes para a COVID-19 (7 dias suavizados). Para países que não relatam dados de testes diariamente, assume-se que os testes mudaram igualmente diariamente em quaisquer períodos em que não foram relatados dados. Isto produz uma série completa de números diários, que é então calculada como média em uma janela rolante de 7 dias;
* **new_tests_smoothed_per_thousand** - Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas;
* **positive_rate** - A parte dos testes COVID-19 que são positivos, dada como média móvel de 7 dias (este é o inverso dos tests_per_case);
* **tests_per_case** - Testes realizados por novo caso confirmado de COVID-19, dado como uma média móvel de 7 dias (este é o inverso da positive_rate);
* **tests_units** - Unidades utilizadas pelo local para relatar seus dados de teste;
* **total_vaccinations** - Número total de doses de vacinação COVID-19 administradas;
* **people_vaccinated** - Número total de pessoas que receberam pelo menos uma dose de vacina;
* **people_fully_vaccinated** - Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação inicial;
* **total_boosters** - Número total de doses de reforço de vacinação COVID-19 administradas (doses administradas além do número prescrito pelo protocolo de vacinação);
* **new_vaccinations** - Novas doses de vacinação COVID-19 administradas (apenas calculadas para dias consecutivos);
* **new_vaccinations_smoothed** - Novas doses de vacinação COVID-19 administradas (7 dias suavizadas). Para países que não reportam dados de vacinação diariamente, assume-se que a vacinação mudou igualmente diariamente durante quaisquer períodos em que não foram reportados dados. Isto produz uma série completa de números diários, que é então calculada como média em uma janela rolante de 7 dias;
* **total_vaccinations_per_hundred** - Número total de doses de vacinação COVID-19 administradas por 100 pessoas na população total;
* **people_vaccinated_per_hundred** - Número total de pessoas que receberam pelo menos uma dose de vacina por 100 pessoas na população total;
* **people_fully_vaccinated_per_hundred** - Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação inicial por 100 pessoas na população total;
* **total_boosters_per_hundred** - Número total de doses de reforço da vacinação COVID-19 administradas por 100 pessoas na população total;
* **new_vaccinations_smoothed_per_million** - Novas doses de vacinação COVID-19 administradas (7 dias suavizadas) por 1.000.000 de pessoas na população total;
* **new_people_vaccinated_smoothed** - Número diário de pessoas que recebem sua primeira dose de vacina (7 dias suavizados);
* **new_people_vaccinated_smoothed_per_hundred** - Número diário de pessoas que recebem sua primeira dose de vacina (7 dias suavizados) por 100 pessoas na população total;
* **stringency_index** - Índice de Estringência de Resposta do Governo: medida composta baseada em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita);
* **population** - População (últimos valores disponíveis). Veja https://github.com/owid/covid-19-data/blob/master/scripts/input/un/population_latest.csv para a lista completa das fontes;
* **population_density** - Número de pessoas dividido por área de terreno, medido em quilômetros quadrados, ano mais recente disponível;
* **median_age** - Idade média da população, projeção da ONU para 2020;
* **aged_65_older** - Parte da população com 65 anos de idade ou mais, ano mais recente disponível;
* **aged_70_older** - Participação da população com 70 anos de idade ou mais em 2015;
* **gdp_per_capita** - Produto interno bruto em paridade de poder de compra (dólar internacional constante em 2011), ano mais recente disponível;
* **extreme_poverty** - Participação da população vivendo em extrema pobreza, ano mais recente disponível desde 2010;
* **cardiovasc_death_rate** - Taxa de mortalidade por doença cardiovascular em 2017 (número anual de mortes por 100.000 pessoas);
* **diabetes_prevalence** - Prevalência do diabetes (% da população de 20 a 79 anos de idade) em 2017;
* **female_smokers** - Participação de mulheres que fumam, ano mais recente disponível;
* **male_smokers** - Participação de homens que fumam, ano mais recente disponível;
* **handwashing_facilities** - Parte da população com instalações básicas de lavagem de mãos nas instalações, o ano mais recente disponível;
* **hospital_beds_per_thousand** - Leitos hospitalares por 1.000 pessoas, o ano mais recente disponível desde 2010;
* **life_expectancy** - Expectativa de vida ao nascer em 2019;
* **human_development_index** - Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano - uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019, importados de http://hdr.undp.org/en/indicators/137506;
* **excess_mortality** - Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020-2021 e o número projetado de mortes para o mesmo período, com base nos anos anteriores. Para mais informações, consulte https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality;
* **excess_mortality_cumulative** - Diferença percentual entre o número cumulativo de mortes desde 1 de janeiro de 2020 e as mortes acumuladas projetadas para o mesmo período, com base nos anos anteriores. Para mais informações, consulte https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality;
* **excess_mortality_cumulative_absolute** - Diferença acumulada entre o número de mortes relatadas desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período, com base nos anos anteriores. Para mais informações, consulte https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality;
* **excess_mortality_cumulative_per_million** - Diferença acumulada entre o número de mortes relatadas desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas. Para mais informações, veja https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality.

Agora que sabemos as últimas datas são do dia 08/08/2022, vamos visualizar os países que mais sofrem com COVID-19 quanto a número de novos casos e número de mortes. No entanto, esse conjunto de dados tem vários agrupamentos na coluna location como, por exemplo, "Europe", "World" e "High income". Isso acaba prejudicando a nossa análise quanto aos resultados dos países. Portanto essas linhas serão removidas abaixo. Isso é possível através da coluna "iso_code", pois esses valores tem o padrão de começar com "OWID".

```
# Verificando os valores únicos da coluna "location"
df_clean.location.unique()
```

```
# Removendo os agrupamentos e mantendo apenas os países
df_countries = df_clean[~df_clean.iso_code.str.contains("OWID*")]
```

```
# Verificar os 5 países que mais sofrem recentemente (08/08/2022) quanto ao número de casos
df_countries.loc[df_countries.date == '2022-08-08'].sort_values(by = "total_cases", ascending = False)[:5]
```

```
# Verificar os 5 países que mais sofrem recentemente (08/08/2022) quanto a número de mortes
df_countries.loc[df_countries.date == '2022-08-08'].sort_values(by = "total_deaths", ascending = False)[:5]
```

```
# Criando conjunto de dados dos 5 países que mais sofreram com casos de COVID-19 no dia 08/08/2022
top5_total_cases = df_countries.loc[df_countries.date == '2022-08-08'].sort_values(by = "total_cases", ascending = False)[:5]
top5_total_cases.reset_index()
```

```
# Plotar os 5 países que mais sofreram com casos de COVID-19 no dia 08/08/2022
fig, ax = plt.subplots(figsize=(10,6))

#top5_total_cases.plot(x= "location", y="total_cases", kind="bar", ax=ax)
ax.bar('location', 'total_cases', data = top5_total_cases)

ax.set_title("Países que mais sofrem com casos de COVID-19 - 08/08/2022")
ax.set_xlabel("País")
ax.set_ylabel("Nº casos")

plt.tight_layout()
```

```
# Criando conjunto de dados dos 5 países que mais sofreram com mortes por COVID-19 no dia 08/08/2022
top5_total_deaths = df_countries.loc[df_countries.date == '2022-08-08'].sort_values(by = "total_deaths", ascending = False)[:5]
top5_total_deaths.reset_index()
```

```
# Plotar os 5 países que mais sofreram com mortes por COVID-19 no dia 08/08/2022
fig, ax = plt.subplots(figsize=(10,6))

#top5_total_cases.plot(x= "location", y="total_cases", kind="bar", ax=ax)
ax.bar('location', 'total_deaths', data = top5_total_deaths)

ax.set_title("Países que mais sofrem com mortes por COVID-19 - 08/08/2022")
ax.set_xlabel("País")
ax.set_ylabel("Nº mortes")
plt.tight_layout()
```

Vamos plotar um gráfico de dispersão considerando `gdp_per_capita` (ou seja, o PIB dos países) pelo número total de mortos.

```
# Plotando gráfico de dispersão
fig, ax = plt.subplots()


df_dispersao = df_countries.loc[df_countries.date == '2022-08-08']
# plotar median_income vs. median_house_value
df_dispersao.plot.scatter(x='total_deaths', y='gdp_per_capita', ax=ax, s=1.0)
ax.set_title("PIB x Nº mortes")
ax.set_xlabel("Nº mortes")
ax.set_ylabel("PIB")

plt.show()
```

Como pode ser visto a pandemia afetou todos os países, porém claramente existe uma concentração nos países mais pobres (onde tem mais bolinhas).

### Análise Exploratória para o Brasil

* Criar uma cópia do DataFrame apenas com o Brasil
    * `df_brasil = df.loc[df.location == "Brazil"].copy()`
* Identificar quando foi registrada a primeira morte no Brasil
* Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
* Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

```
# Criando um conjunto de dados do Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()
df_brasil.head()
```

Precisamos remover os NaN das colunas total_deaths e new_deaths, pois significam que ainda não tinham mortes nessa data. Substituiremos esses valores ausentes por zero.

```
# Substituindo NaN por 0 nas colunas de número de mortes e novas mortes
df_brasil = df_brasil.fillna({"total_deaths": 0})
df_brasil = df_brasil.fillna({"new_deaths": 0})
df_brasil.head()
```

```
# Identificando a primeira morte no Brasil
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)

primeira_morte = df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1).date.iloc[0]

primeira_morte = primeira_morte.strftime("%d/%m/%Y")

print("A primeira morte por COVID-19 no Brasil aconteceu em ", primeira_morte, ".")
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
```

```
# Identificando o primeiro caso no Brasil
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)

primeiro_caso = primeira_morte = df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1).date.iloc[0]

primeiro_caso = primeiro_caso.strftime("%d/%m/%Y")

print("O primeiro caso de COVID-19 no Brasil aconteceu em ", primeiro_caso, ".")
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

```
# Identificando o tempo que levou entre o primeiro caso e a primeira morte
from datetime import date

tempo = df_brasil.date.loc[25877] - df_brasil.date.loc[25857]

print("O tempo entre o primeiro caso e a primeira morte foi de ", tempo.days, "dias.")
```

Vamos agora verificar a evolução dos casos e mortes de COVID-19 no Brasil num gráfico de linhas.

```
# Plotando os gráficos
x = list(range(df_brasil.shape[0]))
y1 = df_brasil.total_cases
y2 = df_brasil.total_deaths

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,6))

ax1.set_title("Casos de COVID-19 - Brasil")
ax1.set_xlabel("Dias")
ax1.set_ylabel("Nº Casos")
ax1.plot (x, y1)

ax2.set_title("Casos de COVID-19 - Brasil (Log)")
ax2.set_xlabel("Dias")
ax2.set_ylabel("Nº Casos")
ax2.set_yscale("log")
ax2.plot (x, y1)

ax3.set_title("Morte por COVID-19 - Brasil")
ax3.set_xlabel("Dias")
ax3.set_ylabel("Nº Mortes")
ax3.plot(x, y2)

ax4.set_title("Mortes por COVID-19 - Brasil (Log)")
ax4.set_xlabel("Dias")
ax4.set_ylabel("Nº Mortes")
ax4.set_yscale("log")
ax4.plot(x, y2)

plt.tight_layout()
```
